The Breast Cancer Wisconsin (Diagnostic) dataset is a widely used benchmark dataset for machine learning algorithms, particularly for classification tasks. It contains data from 569 patients with breast tumors, with each tumor described by 30 numeric features that have been computed from digitized images of a fine needle aspirate (FNA) of the breast mass. These features encapsulate characteristics of the cell nuclei present in the images, including aspects such as radius, texture, perimeter, area, smoothness, compactness, concavity, concave points, symmetry, and fractal dimension.

For each of these characteristics, there are three types of measurements provided in the dataset:

The mean value of the feature across the image.
The standard error of the feature.
The "worst" or largest mean value found among three largest values of the feature.
The primary goal when using this dataset in machine learning is to classify each tumor as either benign (non-cancerous) or malignant (cancerous) based on these features. The 'diagnosis' is the target variable for prediction and is a binary categorical variable represented by 'M' (malignant) and 'B' (benign) in the dataset.

The dataset is lauded for its clear structure, absence of missing values, and the practicality of its application to real-world diagnostic challenges. It is used for education, research, and benchmarking the performance of different machine learning algorithms. Researchers also utilize this dataset to develop and refine algorithms that can assist medical professionals in diagnosing breast cancer, thus aiming to contribute to early detection and improved patient outcomes.

## Correlation matrix:

```python
# The correlation matrix
import matplotlib.pyplot as plt
import seaborn as sns
c=dataset.corr()
plt.figure(figsize=(30,15))
sns.heatmap(c,annot=True,fmt=".2f",cmap="coolwarm",square=True,linewidth=".5",cbar_kws={"shrink":.5})
```

```python
# Print highly correlated features and their correlation values
c = dataset.corr()
upper_triangle = c.where(np.triu(np.ones(c.shape), k=1).astype(bool))
high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.8)]
for feature in high_corr_features:
    correlated_with = upper_triangle.index[upper_triangle[feature] > 0.8].tolist()
    correlation_values = upper_triangle[feature][upper_triangle[feature] > 0.8].tolist()
    for correlated_feature, correlation_value in zip(correlated_with, correlation_values):
        print(f"{feature} is highly correlated with {correlated_feature} (Correlation value: {correlation_value})")
```

## Histograms:

```python
# Exclude the 'diagnosis' column which is categorical
numerical_attributes = dataset.select_dtypes(include=['float64']).columns
num_columns = 4
num_rows = len(numerical_attributes) // num_columns + (len(numerical_attributes) % num_columns > 0)
fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, num_rows * 4))
axes = axes.flatten()
for i, attribute in enumerate(numerical_attributes):
    axes[i].hist(dataset[attribute], bins=20, color='blue', alpha=0.7)
    axes[i].set_title(attribute)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    mean_val = dataset[attribute].mean()
    axes[i].axvline(mean_val, color='red', linestyle='dashed', linewidth=1)
    axes[i].text(mean_val*1.05, axes[i].get_ylim()[1]*0.95, 'Mean: {:.2f}'.format(mean_val))
for i in range(len(numerical_attributes), len(axes)):
    axes[i].axis('off')

plt.tight_layout()

plt.show()
```

## SVM:

```python
# Split the data into features and target variable
X = dataset.drop('diagnosis', axis=1)
y = dataset['diagnosis']
```

```python
# Scale the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Initialize the SVM classifier
classifier = SVC(kernel='linear')
```

```python
# Train the classifier
classifier.fit(X_train, y_train)
```

```python
# Predict the labels of the test set
y_pred = classifier.predict(X_test)
```

```python
# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Hyperparameter tuning using Grid Search
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
grid.fit(X_train, y_train)
```

```python
# Print out the best parameters and the best score
print("Best Parameters: ", grid.best_params_)
print("Best Score: ", grid.best_score_)
```

```python

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

## Logistic Regression

# KNN:

# Random Forrests

```python
# Split the data into features and target variable
X = dataset.drop('diagnosis', axis=1)
y = dataset['diagnosis']

# Scale the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialize the Random Forest classifier
random_forest = RandomForestClassifier(max_features='sqrt')  # Setting max_features explicitly

# Train the classifier
random_forest.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = random_forest.predict(X_test)

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Hyperparameter tuning using Grid Search
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_features': ['sqrt', 'log2'],  # Number of features to consider at every split
    'max_depth': [None, 10, 20, 30, 40, 50],  # Maximum number of levels in tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]  # Minimum number of samples required at each leaf node
}
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=2, cv=5)
grid.fit(X_train, y_train)

# Print out the best parameters and the best score
print("Best Parameters: ", grid.best_params_)
print("Best Score: ", grid.best_score_)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
